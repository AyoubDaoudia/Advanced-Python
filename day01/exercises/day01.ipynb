```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
s_op = [4,6,4,8,6,5,9,9,8,11,13,13,14,13,14,17,16,18,17,18]
a_op = [1,3,3,4,5,7,7,7,10,10,12,12,13,13,13,14,14,14,14,17]

plt.plot(list(range(1,21)),s_op,color="red",label="Without Optimization")
plt.plot(list(range(1,21)),a_op,color="blue",label="With Optimization")
plt.xlabel("Strides")
plt.ylabel("Time (msec)")
plt.legend()
plt.title("CPU runtime comparison with and without optimization")
```

```python
df = pd.read_csv("day01ex02data.csv")
times = df[" time (msec)"][1:]
rates = df[" rate"][1:]
```

```python
plt.plot(list(range(1,21)),times,color="blue",label="CPU Time (msec)")
plt.xlabel("Block Size")
plt.ylabel("CPU Time (msec)")
plt.legend()
plt.title("CPU runtime comparison based on block size matrix multiplication")
```

```python
plt.plot(list(range(1,21)),rates,color="blue",label="Effective bandwidth MB/s")
plt.xlabel("Block Size")
plt.ylabel("Effective bandwidth MB/s")
plt.legend()
plt.title("Effective bandwidth comparison based on block size matrix multiplication")
```
